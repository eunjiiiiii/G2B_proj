```python
import pandas as pd
```

```python
df_g = pd.read_csv('./2019공사공고.csv')
df_t1 =pd.read_csv('./2019공사투찰1.csv')
df_t2 =pd.read_csv('./2019공사투찰2.csv')
df_t3 =pd.read_csv('./2019공사투찰3.csv')
df_t4 =pd.read_csv('./2019공사투찰4.csv')
```

```python
df_t = pd.concat([df_t1, df_t2, df_t3, df_t4])
```

```python
g= [
'입찰공고번호',
 '낙찰업체투찰률',
  '낙찰업체투찰금액',
 '국제입찰여부',
 '정정공고여부',
 '공사현장'
 ,'공공조달분류명'
]

t = [ '입찰서개시일자',
 '입찰서마감일자',
 '조달구분',
 '입찰공고번호',
 '입찰공고차수',
 '수요기관구분',
 '수요기관지역명',
 '추정가격',
 '기초금액',
 '참가수',
 '입찰률',
 '입찰금액',
 '낙찰자결정방법',
'입찰계약방법',
 '공동도급구성방식명',
'개찰순위',
'입찰방식',
'예정가격',
  '낙찰하한율',
  '낙찰자결정적용법규',
]
```

```python
df = pd.merge(df_g[g], df_t[t], how='inner', on=['입찰공고번호'])
```

```python
df.dropna(inplace=True)
```

```python
df['공사현장'] = df['공사현장'].str[:4]
df['공사현장'].loc[df['공사현장'].str.contains('경기')] = '경기도'
df['공사현장'].loc[df['공사현장'].str.contains('전라남도')] = '전라남도'
df['공사현장'].loc[df['공사현장'].str.contains('전라북도')] = '전라북도'
df['공사현장'].loc[df['공사현장'].str.contains('경상남도')] = '경상남도'
df['공사현장'].loc[df['공사현장'].str.contains('경상북도')] = '경상북도'
df['공사현장'].loc[df['공사현장'].str.contains('충청남도')] = '충청남도'
df['공사현장'].loc[df['공사현장'].str.contains('충청북도')] = '충청북도'
df['공사현장'].loc[df['공사현장'].str.contains('강원')] = '강원도'
df['공사현장'].loc[df['공사현장'].str.contains('세종')] = '세종특별시'
df['공사현장'].loc[df['공사현장'].str.contains('대전')] = '대전광역시'
df['공사현장'].loc[df['공사현장'].str.contains('울산')] = '울산광역시'
df['공사현장'].loc[df['공사현장'].str.contains('제주')] = '제주도'
df['공사현장'].loc[df['공사현장'].str.contains('인천')] = '인천광역시'
df['공사현장'].loc[df['공사현장'].str.contains('서울')] = '서울특별시'
df['공사현장'].loc[df['공사현장'].str.contains('부산')] = '부산광역시'
df['공사현장'].loc[df['공사현장'].str.contains('대구')] = '대구광역시'
```

```python
df['수요기관지역명'] = df['수요기관지역명'].str[:4]
df.loc[df['수요기관지역명'] == '서울특별', '수요기관지역명'] = '서울특별시'
df.loc[df['수요기관지역명'] == '대구광역', '수요기관지역명'] = '대구광역시'
df.loc[df['수요기관지역명'] == '대전광역', '수요기관지역명'] = '대전광역시'
df.loc[df['수요기관지역명'] == '부산광역', '수요기관지역명'] = '부산광역시'
df.loc[df['수요기관지역명'] == '울산광역', '수요기관지역명'] = '울산광역시'
df.loc[df['수요기관지역명'] == '광주광역', '수요기관지역명'] = '광주광역시'
df.loc[df['수요기관지역명'] == '인천광역', '수요기관지역명'] = '인천광역시'
df.loc[df['수요기관지역명'] == '제주특별', '수요기관지역명'] = '제주특별시'
df.loc[df['수요기관지역명'] == '세종특별', '수요기관지역명'] = '세종특별시'
```

```python
df.drop(axis = 1,columns = ['입찰서개시일자','입찰서마감일자'],inplace= True)
```

```python
df.to_csv('Final_Gower.csv')
```
